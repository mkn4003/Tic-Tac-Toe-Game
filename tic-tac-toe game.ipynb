```python
def tic_tac_toe_game():

    turn = 'X'
    count = 0
    
    for i in range(1,10):
        printBoard(Board)
        print("\n"+turn + ", It's your turn." + " Move to which place? ")
        move = int(input())
        print()

        if Board[move] == ' ':
            Board[move] = turn
            count += 1
        else:
            print("That place is already filled. \nMove to which place? ")
            continue
                
        # Now we will check if player X or O has won, for every move after 5. 
        if count >= 5:
            if Board[7] == Board[8] == Board[9] != ' ': # across the bottom
                printBoard(Board)
                print("\nGame Over.")                
                print(turn + " won.\n")                
                break
            elif Board[4] == Board[5] == Board[6] != ' ': # across the middle
                printBoard(Board)
                print("\nGame Over.")                
                print(turn + " won.\n")
                break
            elif Board[1] == Board[2] == Board[3] != ' ': # across the top
                printBoard(Board)
                print("\nGame Over.")                
                print(turn + " won.\n")
                break
            elif Board[1] == Board[4] == Board[7] != ' ': # down the left side
                printBoard(Board)
                print("\nGame Over.")                
                print(turn + " won.\n")
                break
            elif Board[2] == Board[5] == Board[8] != ' ': # down the middle
                printBoard(Board)
                print("\nGame Over.")                
                print(turn + " won.\n")
                break
            elif Board[3] == Board[6] == Board[9] != ' ': # down the right side
                printBoard(Board)
                print("\nGame Over.")                
                print(turn + " won.\n")
                break 
            elif Board[7] == Board[5] == Board[3] != ' ': # diagonal
                printBoard(Board)
                print("\nGame Over.")                
                print(turn + " won.\n")
                break
            elif Board[1] == Board[5] == Board[9] != ' ': # diagonal
                printBoard(Board)
                print("\nGame Over.")                
                print(turn + " won.\n")
                break 

        # If neither X nor O wins and the board is full, we'll declare the result as 'tie'.
        if count == 9:
            print("\nGame Over.")                
            print("It's a Tie!!\n")

        # we have to change the player after every move.
        if turn =='X':
            turn = 'O'
        else:
            turn = 'X'
            
    while True:
        restart = input("Do want to play Again?(y/n) : ")
        if restart == "y" or restart == "Y":
            for key in board_keys:
                Board[key] = " "
            tic_tac_toe_game()
        else:
            print("Okay bye, See You Soon :) ")
            break
```

```python
Board = {} # Using Dictionary
for key in range(1,10):
    Board[key] = " "
    
def printBoard(board):
    print(board[1] + '|' + board[2] + '|' + board[3])
    print('-+-+-')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('-+-+-')
    print(board[7] + '|' + board[8] + '|' + board[9])
    
board_keys = []

for key in Board:
    board_keys.append(key)
    
start = input("Do want to play?(y/n): ")
if start == "y" or start == "Y":
    for key in board_keys:
        Board[key] = " "
    tic_tac_toe_game()
```
